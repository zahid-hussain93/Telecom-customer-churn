#**Customer satisfication indicators and churn**

## insight

```python

data = {
    'Online Security': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
    'Online Backup': ['Yes', 'No', 'No', 'Yes', 'Yes', 'No'],
    'Premium Tech Support': ['No', 'Yes', 'No', 'Yes', 'No', 'Yes'],
    'Customer Status': ['Churned', 'Not Churned', 'Churned', 'Not Churned', 'Churned', 'Not Churned']
}

df = pd.DataFrame(data)

# Converting categorical variables to numerical values (Yes/No to 1/0)
satisfaction_columns = ['Online Security', 'Online Backup', 'Premium Tech Support']
for col in satisfaction_columns:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# Group by churn status and calculate the mean satisfaction values
mean_satisfaction = df.groupby('Customer Status')[satisfaction_columns].mean()
```

## visualization

```python
mean_satisfaction.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Customer Satisfaction Indicators for Churned and Not Churned Customers')
plt.xlabel('Customer Status')
plt.ylabel('Mean Satisfaction')
plt.xticks([0, 1], ['Not Churned', 'Churned'], rotation=0)
plt.legend(title='Satisfaction Indicator')
plt.show()
```

# **availability of unlimited data plans or device protection plans impacts churn rates**

## insight

```python

columns_to_analyze = ['Unlimited Data', 'Device Protection Plan', 'Customer Status']

plan_df = df[columns_to_analyze]

# Cleaning and preprocessing
plan_df['Unlimited Data'] = plan_df['Unlimited Data'].map({'Yes': 'With Unlimited Data', 'No': 'Without Unlimited Data'})
plan_df['Device Protection Plan'] = plan_df['Device Protection Plan'].map({'Yes': 'With Device Protection', 'No': 'Without Device Protection'})

# Group by
churn_by_plans = plan_df.groupby(['Unlimited Data', 'Device Protection Plan', 'Customer Status']).size().unstack(fill_value=0)
```

```python
print(churn_by_plans)
```

## visualization

```python
churn_by_plans.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Churn Rates by Unlimited Data and Device Protection Plans')
plt.xlabel('Plan Availability')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.legend(title='Customer Status', loc='upper right')
plt.show()
```
